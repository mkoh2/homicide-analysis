```python
unsolved = df[df['Solved'] == 'No']
print(f"Unsolved cases: {len(unsolved)}")
print(f"\n--- Relationship (unsolved only) ---")
print(unsolved['Relationship'].value_counts())
```

```python
unsolved = df[df['Solved'] == 'No'].copy()
unsolved['VicAge'] = unsolved['VicAge'].replace([0, 999], pd.NA)
print(unsolved['VicAge'].isna().sum(), "unknown ages")
```

```python
weapon_map = {
    'Handgun - pistol, revolver, etc': 'Firearm',
    'Shotgun': 'Firearm',
    'Rifle': 'Firearm',
    'Other gun': 'Firearm',
    'Firearm, type not stated': 'Firearm',
    'Knife or cutting instrument': 'Knife',
    'Blunt object - hammer, club, etc': 'Blunt',
    'Personal weapons, includes beating': 'Personal',
    'Strangulation - hanging': 'Strangulation',
    'Asphyxiation - includes death by gas': 'Strangulation',
    'Fire': 'Other',
    'Drowning': 'Other',
    'Poison - does not include gas': 'Other',
    'Explosives': 'Other',
    'Pushed or thrown out window': 'Other',
    'Narcotics or drugs, sleeping pills': 'Other',
    'Other or type unknown': 'Unknown',
    'Weapon Not Reported': 'Unknown'
}

unsolved['WeaponGroup'] = unsolved['Weapon'].map(weapon_map)
print(unsolved['WeaponGroup'].value_counts())
```

```python
print(unsolved[['Agency', 'State', 'CNTYFIPS']].drop_duplicates().head(20))
```

```python
cities = pd.read_excel('uscities.xlsx')
print(cities.columns.tolist())
print(cities.head(3))
```

```python
# Keep only what we need from cities
cities_slim = cities[['city', 'state_name', 'lat', 'lng']].copy()

# Standardize case for matching
cities_slim['city'] = cities_slim['city'].str.strip().str.title()
cities_slim['state_name'] = cities_slim['state_name'].str.strip()
unsolved['Agency'] = unsolved['Agency'].str.strip().str.title()

# Join on city + state
unsolved_geo = unsolved.merge(
    cities_slim,
    left_on=['Agency', 'State'],
    right_on=['city', 'state_name'],
    how='left'
)

# Check how many matched
matched = unsolved_geo['lat'].notna().sum()
total = len(unsolved_geo)
print(f"Matched: {matched} / {total} ({matched/total*100:.1f}%)")
```

```python
# Parse CNTYFIPS - it looks like "Anchorage, AK" so split on comma
unsolved_geo['county_clean'] = unsolved_geo['CNTYFIPS'].str.split(',').str[0].str.strip().str.title()
unsolved_geo['state_abbr'] = unsolved_geo['CNTYFIPS'].str.split(',').str[1].str.strip()

# Build county centroid lookup from cities file
county_centroids = cities.groupby(['county_name', 'state_id'])[['lat', 'lng']].mean().reset_index()
county_centroids['county_name'] = county_centroids['county_name'].str.strip().str.title()

# For unmatched rows, fill lat/lng from county centroid
unmatched = unsolved_geo['lat'].isna()

unsolved_geo = unsolved_geo.merge(
    county_centroids,
    left_on=['county_clean', 'state_abbr'],
    right_on=['county_name', 'state_id'],
    how='left',
    suffixes=('', '_county')
)

# Fill missing lat/lng with county fallback
unsolved_geo['lat'] = unsolved_geo['lat'].fillna(unsolved_geo['lat_county'])
unsolved_geo['lng'] = unsolved_geo['lng'].fillna(unsolved_geo['lng_county'])

# Final coverage check
matched_final = unsolved_geo['lat'].notna().sum()
print(f"Final matched: {matched_final} / {total} ({matched_final/total*100:.1f}%)")
```

```python
# Drop redundant columns from the joins
unsolved_geo = unsolved_geo.drop(columns=[
    'city', 'state_name', 
    'county_name', 'state_id',
    'lat_county', 'lng_county',
    'county_clean', 'state_abbr'
])

# Verify final shape and check lat/lng look right
print(unsolved_geo.shape)
print(unsolved_geo[['Agency', 'State', 'lat', 'lng']].head(5))
```

```python
def similarity_score(case1, case2):
    score = 0
    
    # Weapon group match (+3)
    if case1['WeaponGroup'] == case2['WeaponGroup'] and case1['WeaponGroup'] != 'Unknown':
        score += 3
    
    # Victim sex match (+2)
    if case1['VicSex'] == case2['VicSex'] and case1['VicSex'] != 'Unknown':
        score += 2
    
    # Victim race match (+1)
    if case1['VicRace'] == case2['VicRace'] and case1['VicRace'] != 'Unknown':
        score += 1
    
    # Victim age within 10 years (+2)
    if pd.notna(case1['VicAge']) and pd.notna(case2['VicAge']):
        if abs(case1['VicAge'] - case2['VicAge']) <= 10:
            score += 2
    
    return score
```

```python
# Grab two cases to test with
case_a = unsolved_geo.iloc[0]
case_b = unsolved_geo.iloc[1]

score = similarity_score(case_a, case_b)
print(f"Score: {score}")
print(f"\nCase A: {case_a['VicSex']}, {case_a['VicRace']}, Age {case_a['VicAge']}, {case_a['WeaponGroup']}")
print(f"Case B: {case_b['VicSex']}, {case_b['VicRace']}, Age {case_b['VicAge']}, {case_b['WeaponGroup']}")
```
