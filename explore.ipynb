```python
import pandas as pd
```

```python
df = pd.read_csv('SHR76_24a.csv', encoding = 'latin-1')
print(df.shape)
```

```python
print(df.columns.tolist())
```

```python
df.head(3)
```

```python
cols_to_check = ['Solved', 'VicSex', 'VicRace', 'Weapon', 'Relationship']

for col in cols_to_check:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())
```

```python
unsolved = df[df['Solved'] == 'No']
print(f"Unsolved cases: {len(unsolved)}")
print(f"\n--- Relationship (unsolved only) ---")
print(unsolved['Relationship'].value_counts())
```

```python
print(unsolved['VicAge'].describe())
print("\nAge value of 0 or 999 (unknown codes):")
print(unsolved['VicAge'].isin([0, 999]).sum())
```
